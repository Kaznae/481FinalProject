```python
import colorsys

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns#import csv data
```

```python
data = pd.read_csv('data/data_all_variables_clean.csv')#define data as dataframe type

df = pd.DataFrame(data)

X = ['2014pm25','Solar_pct','Coal_pct','Oil_pct','Gas_pct','Hydro_pct','Wind_pct','Waste_pct','Biomass_pct','Cogeneration_pct','Other_pct','Petcoke_pct','Nuclear_pct','2013pm25','2014co2']
```

```python
#corr = df[X].corr(method="pearson")

#corr.style.background_gradient(cmap='coolwarm').set_precision(4)fig, ax = plt.subplots(figsize=(12,12))

fig, ax = plt.subplots(figsize=(12,12))

sns.heatmap(df[X].corr(method='pearson'), annot=True, fmt='.4f', cmap=plt.get_cmap('coolwarm'), cbar=False, ax=ax, annot_kws={"size":10})

ax.set_yticklabels(ax.get_yticklabels(), rotation="horizontal")

plt.savefig('corrtablesz12.png')
```
