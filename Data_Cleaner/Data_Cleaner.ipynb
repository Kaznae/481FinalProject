```python
import pandas as pd
import math
import csv
import numpy as np
```

```python
co2 = pd.read_excel('API_EN.ATM.CO2E.PC_DS2_en_excel_v2_10515462.xls')
population = pd.read_excel('API_SP.POP.TOTL_DS2_en_excel_v2_10515202.xls')
```

```python

for i in range(1960, 2018):
    string = str(i)
    if i < 2006 or i > 2014:
        del co2[string]
        del population[string]

print(co2)
```

```python
averages = {}
for index, row in co2.iterrows():
    total = 0
    totalyears = 0
    for i in range(2006, 2015):
        current = str(i)
        prev = str(i-1)
        if math.isnan(row[current]) is False and math.isnan(row[prev]) is False:
            totalyears += 1
            total += (row[current] - row[prev])
    #print(totalyears, ' ', row['Country Name'])
    if totalyears > 0:
        averages [row['Country Code']] = total/totalyears
    else: 
        averages [row['Country Code']] = float('NaN')


print(averages.values())
```

```python
# new_cols = [*list(co2.columns.values),'average C02 increase']
# co2_cleaned = pd.DataFrame(columns = new_cols)

co2 ['average rate of change'] = averages.values()

print(co2)
```

```python
writer = pd.ExcelWriter('cleaned_co2.xlsx')
co2.to_excel(writer,'Sheet5')
writer.save()
```

```python
averages = {}
for index, row in population.iterrows():
    total = 0
    totalyears = 0
    for i in range(2006, 2015):
        current = str(i)
        prev = str(i-1)
        if math.isnan(row[current]) is False and math.isnan(row[prev]) is False:
            totalyears += 1
            total += (row[current] - row[prev])
    #print(totalyears, ' ', row['Country Name'])
    if totalyears > 0:
        averages [row['Country Code']] = round(total/totalyears, 3)
    else: 
        averages [row['Country Code']] = float('NaN')


print(averages.values())
```

```python
population['average population change'] = averages.values()
```
